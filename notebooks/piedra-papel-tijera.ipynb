# Piedra Papel & Tijera

- Reglas del Juego "Piedra, Papel y Tijera":

    - Dos jugadores eligen una de las tres opciones: "Piedra", "Papel" o "Tijera".

    - Las reglas para determinar al ganador son:
        
        - Piedra vence a Tijera.

        - Tijera vence a Papel.

        - Papel vence a Piedra.

        - El jugador que gana una ronda obtiene un punto.
        
        - El juego contin√∫a hasta que un jugador haya acumulado 3 puntos, convirti√©ndose en el ganador.


- Pistas para las alumnas:

    1. Comienza por definir el n√∫mero de puntos objetivo que se deben alcanzar para ganar el juego.

    2. Implementa una funci√≥n que obtenga la elecci√≥n del jugador y valide que sea una de las opciones v√°lidas ("Piedra", "Papel" o "Tijera").

    3. Utiliza un bucle `while` para llevar un registro de los puntos de cada jugador y continuar el juego hasta que se alcance el n√∫mero de puntos objetivo.

    4. Genera una elecci√≥n aleatoria para el segundo jugador (la computadora) utilizando el m√≥dulo `random`.

    5. Determina al ganador de cada ronda y actualiza los puntos correspondientes.

    6. Verifica si alg√∫n jugador ha alcanzado el n√∫mero de puntos objetivo y finalicen el juego cuando eso suceda.

```python
# Importamos
import random

# Definimos el n¬∫ de puntos necesarios para ganar
PUNTOS_OBJETIVO = 3




# Definimos la funci√≥n para obtener y validar la elecci√≥n del jugador
def obtener_eleccion_jugador():
    opciones = ["piedra", "papel", "tijera", "salir"]
    eleccion = input("Elige piedra, papel, tijera o escribe 'salir' para terminar el juego: ").lower()
    while eleccion not in opciones:
        print("Opci√≥n no v√°lida. Intenta de nuevo.")
        eleccion = input("Elige piedra, papel, tijera o 'salir': ").lower()
    return eleccion




# Definimos la funci√≥n para determinar el ganador de la ronda
def determinar_ganador(jugador, ordenador):
    if jugador == ordenador:
        return "empate"
    elif (
        (jugador == "piedra" and ordenador == "tijera") or
        (jugador == "tijera" and ordenador == "papel") or
        (jugador == "papel" and ordenador == "piedra")
    ):
        return "jugador"
    else:
        return "ordenador"




# Generamos funci√≥n principal
def jugar ():
    puntos_jugador = 0
    puntos_ordenador = 0

# Generamos un bucle "while" para llevar un registro de los puntos de cada jugador y continuar el juego hasta que se alcance el n√∫mero de puntos objetivo.
    print("\n--- ¬°Comenzamos! ---")

    while puntos_jugador < PUNTOS_OBJETIVO and puntos_ordenador < PUNTOS_OBJETIVO:
        print("\n--- Nueva ronda ---")
        print(f"Puntos -> Jugador: {puntos_jugador} | Ordenador: {puntos_ordenador}")
    
        jugador = obtener_eleccion_jugador()
    
        if jugador == "salir":
            print("Has salido del juego. ¬°Hasta la pr√≥xima!")
            break

        ordenador = random.choice(["piedra", "papel", "tijera"])
        print(f"El ordenador eligi√≥: {ordenador}")

        ganador = determinar_ganador(jugador, ordenador)

        if ganador == "jugador":
            print("¬°Ganaste esta ronda!")
            puntos_jugador += 1

        elif ganador == "ordenador":
            print("El ordenador gana esta ronda.")
            puntos_ordenador += 1

        else:
            print("¬°Empate!")

# Mostramos el resultado final si no se sali√≥ manualmente
    if puntos_jugador == PUNTOS_OBJETIVO:
            print("\n=== Juego terminado ===")
            print(
            """
‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà ‚ñà‚ñà      ‚ñà‚ñà  ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà ‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà   ‚ñà‚ñà‚ñà‚ñà‚ñà  ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà  ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà 
‚ñà‚ñà      ‚ñà‚ñà      ‚ñà‚ñà      ‚ñà‚ñà ‚ñà‚ñà      ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà ‚ñà‚ñà      ‚ñà‚ñà      
‚ñà‚ñà‚ñà‚ñà‚ñà   ‚ñà‚ñà‚ñà‚ñà‚ñà   ‚ñà‚ñà      ‚ñà‚ñà ‚ñà‚ñà      ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà   ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà 
‚ñà‚ñà      ‚ñà‚ñà      ‚ñà‚ñà      ‚ñà‚ñà ‚ñà‚ñà      ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà ‚ñà‚ñà           ‚ñà‚ñà 
‚ñà‚ñà      ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà ‚ñà‚ñà  ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà ‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà  ‚ñà‚ñà   ‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà  ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà                                                                        
            """)
            print("Ganaste el juego. üéâ")

    if puntos_ordenador == PUNTOS_OBJETIVO:
            print("\n=== Juego terminado ===")
            print(
            """
 ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà  ‚ñà‚ñà   ‚ñà‚ñà        ‚ñà‚ñà    ‚ñà‚ñà  ‚ñà‚ñà‚ñà‚ñà‚ñà  ‚ñà‚ñà    ‚ñà‚ñà  ‚ñà‚ñà‚ñà‚ñà‚ñà  
‚ñà‚ñà    ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà        ‚ñà‚ñà    ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà  ‚ñà‚ñà  ‚ñà‚ñà  ‚ñà‚ñà   ‚ñà‚ñà 
‚ñà‚ñà    ‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà        ‚ñà‚ñà    ‚ñà‚ñà ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà   ‚ñà‚ñà‚ñà‚ñà   ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà 
‚ñà‚ñà    ‚ñà‚ñà ‚ñà‚ñà   ‚ñà‚ñà         ‚ñà‚ñà  ‚ñà‚ñà  ‚ñà‚ñà   ‚ñà‚ñà    ‚ñà‚ñà    ‚ñà‚ñà   ‚ñà‚ñà 
 ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà  ‚ñà‚ñà   ‚ñà‚ñà ‚ñÑ‚ñà       ‚ñà‚ñà‚ñà‚ñà   ‚ñà‚ñà   ‚ñà‚ñà    ‚ñà‚ñà    ‚ñà‚ñà   ‚ñà‚ñà                                                                                                                                                                                      
            """)
            print("El ordenador gan√≥ el juego. üò¢")


# Llamamos a la funci√≥n
jugar()
```
